```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
k_gauss = 15 # must be positive and odd
k_lapl = 21
```

```python
def display(img):
    plt.rcParams['figure.figsize'] = [10, 10]
    plt.imshow(img, cmap='gray')
    plt.axis('off')
```

```python
im = cv2.imread('flower.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
display(im)
```

```python
im_gauss = cv2.GaussianBlur(im, (k_gauss,k_gauss), 0)
display(im_gauss)
```

```python
im_lapl = cv2.Laplacian(im,cv2.CV_64F)
display(im_lapl)
```

# Laplaciana da Gaussiana

```python
im_gauss_lapl = cv2.Laplacian(im_gauss,cv2.CV_64F, ksize=k_lapl)
# https://stackoverflow.com/questions/2521704/laplacian-of-gaussian-how-does-it-work-opencv
# resulting image is already scaled in (0,1) values
display(im_gauss_lapl)
```
